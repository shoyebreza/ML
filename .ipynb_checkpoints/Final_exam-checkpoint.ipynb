```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

employee = pd.read_csv('employee.csv')
employee
```

## Question : 1

```python
TrainingHours = np.array(employee['TrainingHours'])
TrainingHours.astype(np.float32)
mean_val = np.mean(TrainingHours)
std_val = np.std(TrainingHours)
print(mean_val)
print(std_val)
```

## Question : 2

```python

employee['ExperienceYears'].fillna(employee['ExperienceYears'].mean(), inplace=True)

experience_employees = employee[(employee['ExperienceYears'] > 12) & (employee['ProjectCount'] > 5)]
total_salary = experience_employees['Salary'].sum()

print(total_salary)
```

## Question : 3

```python
filtered_emp = employee[((employee['Department'] == "IT") | (employee['Department'] == "Finance")) & ((employee['Age'] <= 45) | (employee['Age'] >= 30)) & (employee['PerformanceScore'] >= 80)]
sorted_emp = filtered_emp.sort_values(by=['Salary'], ascending=False)
final_col = sorted_emp[['Name', 'Department','Salary','PerformanceScore']]
final_col
```

## Question : 5

```python
import plotly.express as px
counts = employee.groupby(['Department', 'Gender']).size().reset_index(name='Count')

fig = px.bar(counts, x='Department', y='Count', color='Gender',
    barmode='group',title='Employee Count Department and Gender',
    labels={'Department': 'Department', 'Count': 'Employees'})

fig.show()
```

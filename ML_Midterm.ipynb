```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

## Q3. answer

```python
import numpy as np

emp_ages = np.array([56, 46, 32, 25, 38, 56, 36, 40, 28, 28])
age_greater_30 = emp_ages[emp_ages > 30]
print("Ages greater than 30:", age_greater_30)
```

```python
total_greater_30 = np.count_nonzero(age_greater_30)
print("Total of ages greater than 30:", total_greater_30)
```

## Q5. answer

```python
import numpy as np

salaries = np.array([77191, 74131, 46023, 71090, 31685, 30769, 89735, 86101, 32433, 35311])
total_salary_expense = np.sum(salaries)
print("Total Salary Expense:", total_salary_expense)

min_salary = np.min(salaries)
max_salary = np.max(salaries)

differences = max_salary - min_salary
print("Difference between the highest and lowest salaries:", differences)
```

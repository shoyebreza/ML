```python
import pandas as pd

# Create a 4x3 DataFrame
data = {'Student': ['Alice', 'Bob', 'Charlie', 'David'],
        'Math': [90, 85, 92, 88],
        'Science': [88, 91, 85, 90]}
df_students = pd.DataFrame(data)
print("Original DataFrame:")
print(df_students)

# Drop a column (e.g., 'Science')
df_students_dropped_col = df_students.drop('Science', axis=1)
print("\nDataFrame after dropping 'Science' column:")
print(df_students_dropped_col)

# Drop a row by index (e.g., index 2)
df_students_dropped_row = df_students.drop(2, axis=0)
print("\nDataFrame after dropping row with index 2:")
print(df_students_dropped_row)
```

```python
import pandas as pd

# Create a DataFrame of employees
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Alice'],
        'Department': ['HR', 'Engineering', 'HR', 'Sales', 'Engineering'],
        'Salary': [50000, 80000, 55000, 60000, 75000]}
df_employees = pd.DataFrame(data)
print("Original DataFrame:")
display(df_employees)

# Sort by "Salary" in ascending order and by "Name" in descending order
df_sorted = df_employees.sort_values(by=['Salary', 'Name'], ascending=[True, False])
print("\nDataFrame sorted by Salary (ascending) and Name (descending):")
display(df_sorted)
```

```python
import pandas as pd

# Create a DataFrame of students
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi'],
        'Score': [85, 30, 92, 65, 45, 78, 90, 35]}
df_scores = pd.DataFrame(data)
print("Original DataFrame:")
display(df_scores)

# Filter students with Score > 60
df_high_scores = df_scores[df_scores['Score'] > 60]
print("\nStudents with Score > 60:")
display(df_high_scores)

# Filter students with Score < 40
df_low_scores = df_scores[df_scores['Score'] < 40]
print("\nStudents with Score < 40:")
display(df_low_scores)
```

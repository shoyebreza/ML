```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('practiceday15.csv') 
df
```

```python
df['discount_price'] = df['Price'] * 0.10
df
```

```python
df.to_csv('discounted_products.csv', index=False)
```

```python
duplicate_count = df['EmployeeID'].duplicated().sum()
print('Total duplicate',duplicate_count)
```

```python
remove_duplicate = df.drop_duplicates(subset = 'EmployeeID', keep='first')
remove_duplicate
```

```python
miss_age = df['Age'].isnull().sum() 
missing_salary = df['Salary'].isnull().sum() 

print('missing age', miss_age)
print('Missing salary', missing_salary)
```

```python
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
df
```

```python
df = df.dropna(subset= ['Salary'])
df
```

```python
print("Mean of Marks1:", df['Marks1'].mean())
print("Mean of Marks2:", df['Marks2'].mean())
print("Mean of Marks3:", df['Marks3'].mean())

print("Max of Marks1:", df['Marks1'].max())
print("Max of Marks2:", df['Marks2'].max())
print("Max of Marks3:", df['Marks3'].max())
```

```python
df = df.copy()
df['Total_Marks'] = df['Marks1'] + df['Marks2'] + df['Marks3']

print(df[['EmployeeID','EmployeeName','Total_Marks']])
```

```python
std_total = df['Total_Marks'].std()
print("Standard Deviation of Total Marks:", std_total)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


plt.hist(df['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Employee Age')
plt.show()
```
